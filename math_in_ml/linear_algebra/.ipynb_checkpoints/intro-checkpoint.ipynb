#### In theory

1. Vector: 
    -> operations: + (two vectors), * (by scalar);
    -> properities: associative => (a+b)+c = a+(b+c);
    -> vector lenght: |a| = âˆš(a^2+b^2);
    -> dot product (multiplication two vectors): a.b = ai * bi + aj * bj:
        - commutative: a.b = b.a,
        - distributive over addition: a.(b+c) = a.b + a + c,
        - associative over scalar multiplication: a.(Lb) = L(a.b),
        - in orthogonal case: if cos90(degree) = 0(rad) => cos0 = 1 => a.b = |a| * |b|,
        - vector projection: a * a.b / |a| * |a|,
        - scalar projection: a.b / |r|;
        
2. Basis:
    -> set on n vectors:
        - this vector span the space,
        - each vectors are linear independence;
        
3. Matrix:
    -> operations on matrix: 
        - add one row to second row,
        - swap places of two rows,
        - multiplication some row by scalar;
    -> operation between two matrix:
        - addiction,
        - substraction;
        - multiplication (use Einstein summation convention -> abik = sum(aij * bjk), AB = C => cik = aij * bjk);
     -> transform type:
        - shears,
        - invert,
        - mirror,
        - rotation,
        - transpose (Aij.T = Aji),
        - stretch;
      -> type of matrix:
        - zero,
        - orthogonal,
        - row,
        - column,
        - diagonal,
        - unit;
      -> property:
        - determinant,
        - A.T * A = I if |A| = +- 1,
        - A.T = A^(-1)
      * to solve systems of equations we could use Gaussian elimination.

```python
# Vectors intro
vector_1 = [1, 2, 3, 4, 5]

vector_1[0] # 1

vector_1[-1] # 5

len(vector_1) # 5

vector_1[0] + vector_1[1] # 3

vector_1[0]*2 # 2

vector_1.append(6)

for x in vector_1:
    print(x) # 1 2 3 4 5 6
    pass
    
vector_1.remove(2)

for x in vector_1:
    print(x) # 1 3 4 5 6
    pass
    
vector_2 = [5,4,3,2,1]

vector_1[3] + vector_2[0] # 10

vector_1 + vector_2 # [1,3,4,5,6,5,4,3,2,1]
```
