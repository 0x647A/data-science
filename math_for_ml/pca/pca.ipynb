### PCA for high-dimensional datasets

Principal component analysis is mathematical (statistical) procedure uses an orthogonal transformation to change correlated observations into set of values linear uncorrelated. Often you can hear that it is dimensionality reduction. 

For this approach we must know something about variance and covariance matrix.

PCA's steps:
    1. Compute the mean of the data matrix
    2. Substract mean from each data point
    3. Divide the data by its standard deviation in each dimension
    4. Compute the eigenvectors (orthonormal) and eigenvalues
    5. Choose the eigenvectors associated with the M largest eigenvalues to be the basis of the principal  subspace and collect these eigenvectors in a matrix
    6. Orthogonal projection of the data onto the principal axis using the projection matrix

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normalize(X):
    """Normalize the given dataset X
    Args:
        X: ndarray, dataset
    
    Returns:
        (Xbar, mean, std): ndarray, Xbar is the normalized dataset
        with mean 0 and standard deviation 1; mean and std are the 
        mean and standard deviation respectively.
    """
    mu = np.mean(X, axis = 0)
    std = np.std(X, axis=0)
    std_filled = std.copy()
    std_filled[std==0] = 1.
    Xbar = (X-mu) / std_filled
    return Xbar, mu, std

def eig(S):
    """Compute the eigenvalues and corresponding eigenvectors 
        for the covariance matrix S.
    Args:
        S: ndarray, covariance matrix
    
    Returns:
        (eigvals, eigvecs): ndarray, the eigenvalues and eigenvectors
    """
    eigvals, eigvecs = np.linalg.eig(S)
    sort = np.argsort(eigvals)[::-1]
    eigvals = eigvals[sort]
    eigvecs = eigvecs[:,sort]
    return (eigvals, eigvecs)

def projection_matrix(B):
    """Compute the projection matrix onto the space spanned by `B`
    Args:
        B: ndarray of dimension (D, M), the basis for the subspace
    
    Returns:
        P: the projection matrix
    """
    P = B @ np.linalg.inv(np.transpose(B) @ B) @ np.transpose(B)
    return P

# loss function
def mse(predict, actual):
    return np.square(predict - actual).sum(axis=1).mean()
```

```python
def PCA(X, num_components):
    """
    Args:
        X: ndarray of size (N, D), where D is the dimension of the data,
           and N is the number of datapoints
        num_components: the number of principal components to use.
    Returns:
        X_reconstruct: ndarray of size (N, D) the reconstruction
        of X from the first `num_components` principal components.
    """
    # compute the data covariance matrix S
    S = np.cov(X, rowvar = False)

    # find eigenvalues and corresponding eigenvectors for S by implementing eig().
    eig_vals, eig_vecs = eig(S)
    
    # the projection_matrix 
    P = projection_matrix(eig_vecs[:, 0:num_components])
    
    # then for each data point x_i in the dataset X can project the original x_i onto the eigenbasis.
    X_reconstruct = (P @ X.T).T
    return X_reconstruct
```

```python
# how it works?
from keras.datasets import mnist 
(X_train, y_train), (X_test, y_test) = mnist.load_data()


NUM_DATAPOINTS = 1000
X_train = (X_train.reshape(-1, 28 * 28)[:NUM_DATAPOINTS]) / 255.
Xbar, mu, std = normalize(X_train)

loss = []
reconstructions = []
for num_component in range(1, 100):
    reconst = PCA(Xbar, num_component)
    error = mse(reconst, Xbar)
    reconstructions.append(reconst)
    print('n = {:d}, reconstruction_error = {:f}'.format(num_component, error))
    loss.append((num_component, error))

reconstructions = np.asarray(reconstructions)
reconstructions = reconstructions * std + mu 
loss = np.asarray(loss)
```

```python
fig, ax = plt.subplots()
ax.plot(loss[:,0], loss[:,1]);
ax.axhline(100, linestyle='--', color='r', linewidth=2)
ax.xaxis.set_ticks(np.arange(1, 100, 5));
ax.set(xlabel='num_components', ylabel='MSE', title='MSE vs number of PCA');
```

```python
def PCA_high_dim(X, num_components):
    """Compute PCA for small sample size. 
    Args:
        X: ndarray of size (N, D), where D is the dimension of the data,
           and N is the number of data points in the training set. You may assume the input 
           has been normalized.
        num_components: the number of principal components to use.
    Returns:
        X_reconstruct: (N, D) ndarray. the reconstruction
        of X from the first `num_components` principal components.
    """
    N, D = X.shape
    M = (1/N)*(X@X.T) # compute the matrix
    eig_vals, eig_vecs = eig(M) # compute the eigenvalues. 
    U = (X.T @ eig_vecs)[:,:num_components] # compute the eigenvectors for the original PCA problem.
    P = projection_matrix(U) # compute the projection matrix
    X_reconstruct = (P @ X.T).T # compute the projection
    return X_reconstruct
```

```python
# testing
np.testing.assert_almost_equal(PCA(Xbar, 2), PCA_high_dim(Xbar, 2))
print('correct')
```
