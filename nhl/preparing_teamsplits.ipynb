```python
df_teamsplits = df_teamsplits[(df_teamsplits['lgID'] == 'NHL') & (df_teamsplits['year'] >= 1980)]
```

```python
columns_to_delete = df_teamsplits.columns[3:11]
df_teamsplits = df_teamsplits.drop(columns=columns_to_delete)
df_teamsplits = df_teamsplits.drop(columns = 'lgID')
```

```python
df_teamsplits.loc[:, 'tmID'] = pd.Categorical(df_teamsplits['tmID'])
```

```python
df_teamsplits.to_pickle('pickle/teamsplits.pickle')
```
